```python
import os as os
import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np


abnormal_data=[]
normal_data=[]
path_parent = os.path.dirname(os.getcwd())
# taking 10000 data sample for performance and testing

#read file loop and condition
# total file 111423
for i in range(1,111423) :
    if(i<10):
        file_name = path_parent+"/Data_Sample/metabolites/hmdb_metabolites-{:02d}.xml".format(i)
    else:
        file_name = path_parent+"/Data_Sample/metabolites/hmdb_metabolites-{:02d}.xml".format(i)
    tree = ET.parse(file_name)
    root=tree.getroot()
    file=file_name[12:-4]
    
    value=0
    age=""
    sex=""
    info=""
#     comment="None"
    accession= root.find('accession').text
    name= root.find('name').text
    

    
    #print("Abnormal Concentration\n ")
    for abnormal in root.findall('abnormal_concentrations'):

        for concentration in abnormal.findall('concentration'):
    #filter for blood specimen
            if(concentration.find('biospecimen').text=="Blood"):
                for attribute in concentration :
#                     print(attribute.tag,':',attribute.text)
                    if(attribute.tag=='concentration_value'):
                        value=attribute.text
                    if(attribute.tag=='patient_age'):
                        age=attribute.text
                    if(attribute.tag=='patient_sex'):
                        sex=attribute.text
                    if(attribute.tag=='patient_information'):
                        info=attribute.text
#                     if(attribute.tag=='comment'):
#                         comment=attribute.text
                        
    #append array for creating dataframe
                    toappend=[name,accession,value,age,sex,info]
                abnormal_data.append(toappend)



    #print reference(optional)                      
    #                     else : 
    #                         for reference in attribute.findall('reference'):
    #                             print(reference.tag," : ",reference.text)


    for normal in root.findall('normal_concentrations'):
        for concentration in normal.findall('concentration'):
            if(concentration.find('biospecimen').text=="Blood"):
                for attribute in concentration:
                        if(attribute.tag=='concentration_value'):
                            value=attribute.text
                        if(attribute.tag=='subject_age'):
                            age=attribute.text
                        if(attribute.tag=='subject_sex'):
                            sex=attribute.text
                        if(attribute.tag=='subject_condition'):
                            info=attribute.text
#                         if(attribute.tag=='comment'):
#                             comment=attribute.text

                        toappend=[name,accession,value,age,sex,info]
                normal_data.append(toappend)

# create dataframe and write file in csv format   

# abnormal data
abndf = pd.DataFrame(abnormal_data,columns=['Value(uM)','Patient Age','Patient Sex','Patient Information','Name','Accession'])                    
abndf.to_csv(path_parent+"/Output/'Abnormal_Concentration.csv',header='False')
abndf.head(20)
# normal data
ndf = pd.DataFrame(normal_data,columns=['Value(uM)','Subject Age','Subject Sex','Subject Information','Name','Accession'])                    
ndf.to_csv(path_parent+"/Output/'Normal_Concentration.csv',header='False')
# ndf.tail(50)
                    

         
```
