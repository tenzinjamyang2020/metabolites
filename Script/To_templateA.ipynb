```python
import os as os
import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np



def GenerateXML(file):
        creation_date = ""
        update_date=""
        accession=""
        name=""
        
        for get_create in root.findall('creation_date'):
            creation_date = get_create.text
        for get_update in root.findall('update_date'):
            update_date = get_update.text
        for get_accession in root.findall('accession'):
            accession = get_accession.text
        for get_name in root.findall('name'):
            name = get_name.text
            
        to_root=ET.Element("Metabolites")
        c1=ET.Element("creation_date")
        c2=ET.Element("update_date")
        c3=ET.Element("accession")
        c4=ET.Element("normal_concentrations")
        c5=ET.Element("abnormal_concentrations")
        c6=ET.Element("diseases")
        c7=ET.Element("name")
        
        to_root.append(c1)
        to_root.append(c2)
        to_root.append(c3)
        to_root.append(c7)
        to_root.append(c4)
        to_root.append(c5)
        to_root.append(c6)
        
        c1.text=creation_date
        c2.text=update_date
        c3.text=accession
        c7.text=name
        
        for normal in root.findall('normal_concentrations'):   
           
            for concentration_from in normal.findall('concentration'):
                
                if(concentration_from.find('biospecimen').text=="Blood"):
                    concentration=ET.SubElement(c4,"concentration")
                    for attribute in concentration_from :
                        if(attribute.tag=='biospecimen'):
                            biospecimen = ET.SubElement(concentration,"biospecimen")
                            biospecimen.text = attribute.text
                        if(attribute.tag=='concentration_value'):
                            concentration_value = ET.SubElement(concentration,"concentration_value")
                            concentration_value.text = attribute.text
                        if(attribute.tag=='subject_age'):
                            subject_age = ET.SubElement(concentration,"subject_age")
                            subject_age.text = attribute.text
                        if(attribute.tag=='subject_sex'):
                            subject_sex = ET.SubElement(concentration,"subject_sex")
                            subject_sex.text = attribute.text
                        if(attribute.tag=='subject_condition'):
                            subject_condition = ET.SubElement(concentration,"subject_condition")
                            subject_condition.text = attribute.text
                        if(attribute.tag=='comment'):
                            comment = ET.SubElement(concentration,"comment")
                            comment.text = attribute.text
                            
                        if(attribute.tag=='references'):
                            references = ET.SubElement(concentration,"references")
                            for references_from in attribute.findall('reference'):
                                reference=ET.SubElement(references,"reference")
                                for reference_from in references_from: 
                                    if(reference_from.tag=='reference_text'):
                                        reference_text = ET.SubElement(reference,"reference_text")
                                        reference_text.text=reference_from.text
                                    if(reference_from.tag=='pubmed_id'):
                                        pubmed_id = ET.SubElement(reference,"pubmed_id")
                                        pubmed_id.text=reference_from.text
        
        for abnormal in root.findall('abnormal_concentrations'):   
           
            for concentration_from in abnormal.findall('concentration'):

                if(concentration_from.find('biospecimen').text=="Blood"):
                    concentration=ET.SubElement(c5,"concentration")
                    for attribute in concentration_from :
                        if(attribute.tag=='biospecimen'):
                            biospecimen = ET.SubElement(concentration,"biospecimen")
                            biospecimen.text = attribute.text
                        if(attribute.tag=='concentration_value'):
                            concentration_value = ET.SubElement(concentration,"concentration_value")
                            concentration_value.text = attribute.text
                        if(attribute.tag=='patient_age'):
                            patient_age = ET.SubElement(concentration,"patient_age")
                            patient_age.text = attribute.text
                        if(attribute.tag=='patient_sex'):
                            patient_sex = ET.SubElement(concentration,"patient_sex")
                            patient_sex.text = attribute.text
                        if(attribute.tag=='patient_information'):
                            patient_information = ET.SubElement(concentration,"patient_information")
                            patient_information.text = attribute.text
                        if(attribute.tag=='comment'):
                            comment = ET.SubElement(concentration,"comment")
                            comment.text = attribute.text
                            
                        if(attribute.tag=='references'):
                            references = ET.SubElement(concentration,"references")
                            for references_from in attribute.findall('reference'):
                                reference=ET.SubElement(references,"reference")
                                for reference_from in references_from: 
                                    if(reference_from.tag=='reference_text'):
                                        reference_text = ET.SubElement(reference,"reference_text")
                                        reference_text.text=reference_from.text
                                    if(reference_from.tag=='pubmed_id'):
                                        pubmed_id = ET.SubElement(reference,"pubmed_id")
                                        pubmed_id.text=reference_from.text
                                
        
        
        for diseases in root.findall('diseases'):
            for disease_from in diseases:
                disease=ET.SubElement(c6,"disease")
                for attribute in disease_from:
                    if(attribute.tag=='name'):
                        disease_name = ET.SubElement(disease,"name")
                        disease_name.text=attribute.text
#                         print(attribute.text)
                    if(attribute.tag=='omim_id'):
                        omim_id = ET.SubElement(disease,"omim_id")
                        omim_id.text=attribute.text
                    if(attribute.tag=='references'):
                        references = ET.SubElement(disease,"references")
                        for references_from in attribute.findall('reference'):
                            reference=ET.SubElement(references,"reference")
                            for reference_from in references_from: 
                                if(reference_from.tag=='reference_text'):
                                    reference_text = ET.SubElement(reference,"reference_text")
                                    reference_text.text=reference_from.text
                                if(reference_from.tag=='pubmed_id'):
                                    pubmed_id = ET.SubElement(reference,"pubmed_id")
                                    pubmed_id.text=reference_from.text

        to_tree=ET.ElementTree(to_root)
        
        
        with open(file,"wb") as files:
            to_tree.write(files)


for i in range(7,12) :
    
    path_parent = os.path.dirname(os.getcwd())
#     os.chdir(path_parent)
#     print(path_parent)
    if(i<10):
        file_name =path_parent+"/Data_Sample/metabolites/hmdb_metabolites-{:02d}.xml".format(i)
        print(file_name)
    else:
        file_name = path_parent+"/Data_Sample/metabolites/hmdb_metabolites-{}.xml".format(i)
    tree = ET.parse(file_name)
    root=tree.getroot()
    file=file_name[12:-4]
        

    
    GenerateXML(path_parent+"/Output/XML-TemplateA/hmdb_metabolites-{}.xml".format(i))
        
        
#     //update_date , accession , name , normal concentration , 
```
