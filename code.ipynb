```python
import os as os
import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np


abnormal_data=[]
normal_data=[]

# taking 10000 data sample for performance and testing

#read file loop and condition
for i in range(10000) :
    if(i<10):
        file_name = "metabolites/hmdb_metabolites-{:02d}.xml".format(i)
    else:
        file_name = "metabolites/hmdb_metabolites-{}.xml".format(i)
    tree = ET.parse(file_name)
    root=tree.getroot()
    file=file_name[12:-4]


    #print("Abnormal Concentration\n ")
    for abnormal in root.findall('abnormal_concentrations'):

        for concentration in abnormal.findall('concentration'):
    #filter for blood specimen
            if(concentration.find('biospecimen').text=="Blood"):
                for attribute in concentration :
                    if(attribute.tag=='concentration_value'):
                        value=attribute.text
                    if(attribute.tag=='patient_age'):
                        age=attribute.text
                    if(attribute.tag=='patient_sex'):
                        sex=attribute.text
                    if(attribute.tag=='patient_information'):
                        info=attribute.text
                    if(attribute.tag=='comment'):
                        comment=attribute.text
                    else:
                        comment="None"
    #append array for creating dataframe
                    toappend=[value,age,sex,info,comment,file]
                abnormal_data.append(toappend)



    #print reference(optional)                      
    #                     else : 
    #                         for reference in attribute.findall('reference'):
    #                             print(reference.tag," : ",reference.text)


    for normal in root.findall('normal_concentrations'):
        for concentration in normal.findall('concentration'):
            if(concentration.find('biospecimen').text=="Blood"):
                for attribute in concentration:
                        if(attribute.tag=='concentration_value'):
                            value=attribute.text
                        if(attribute.tag=='subject_age'):
                            age=attribute.text
                        if(attribute.tag=='subject_sex'):
                            sex=attribute.text
                        if(attribute.tag=='subject_condition'):
                            info=attribute.text
                        if(attribute.tag=='comment'):
                            comment=attribute.text
                        else:
                            comment="None"
                        toappend=[value,age,sex,info,comment,file]
                normal_data.append(toappend)

# create dataframe and write file in csv format   

# abnormal data
abndf = pd.DataFrame(abnormal_data,columns=['Value(uM)','Patient Age','Patient Sex','Patient Information','Comment','File'])                    
abndf.to_csv('Abnormal_Concentration.csv',header='False')
# abndf.head(20)
# normal data
ndf = pd.DataFrame(normal_data,columns=['Value(uM)','Subject Age','Subject Sex','Subject Information','Comment','File'])                    
ndf.to_csv('Normal_Concentration.csv',header='False')
ndf.tail(50)
                    

         
```
