```python
import os as os
# import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np


# read 10-100000 xml sample
for i in range(0,100) :
    #read 0-10 xml sample (adding zero in front since xml file name start with 0 eg hmdb_metabolites-04)
    if(i<10):
        file_name = "metabolites/hmdb_metabolites-{:02d}.xml".format(i)
    else:
        file_name = "metabolites/hmdb_metabolites-{}.xml".format(i)
    print(file_name)
    tree = ET.parse(file_name)
    root=tree.getroot()


    print("Abnormal Concentration\n ")
    for abnormal in root.findall('abnormal_concentrations'):
        for concentration in abnormal.findall('concentration'):
            print("\n")
            for attribute in concentration :
#                 print(concentration.find('biospecimen').text)
                if(attribute.tag!='references') : 
                    print(attribute.tag," : ",attribute.text)
# push into array
                else : 
                    for reference in attribute.find('reference'):
                        print(reference.tag," : ",reference.text)




    print("========================================================================================================\nNormal Concentration\n ")
    for normal in root.findall('normal_concentrations'):
        for concentration in abnormal.findall('concentration'):
            print("\n")
            for attribute in concentration:
#                 print(concentration.find('biospecimen').text)
              
                if(attribute.tag!='references') : 
                    print(attribute.tag," : ",attribute.text)
                else : 
                    for reference in attribute.find('reference'):
                        print(reference.tag," : ",reference.text)
               
# use numpy to put in array

# create dataframe 

# write file in csv based on template
         
```
